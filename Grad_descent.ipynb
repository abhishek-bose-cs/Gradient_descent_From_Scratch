```python
#Implement Gradient Descent from scratch
import numpy as np
import matplotlib.pyplot as plt

def cal_cost(m,theta,X,Y):
    pred = np.dot(X,theta)
    cost = (1/(2*m))*np.sum(np.square(Y-pred))
    return cost


def grad_desc(m,theta,X,Y,alpha=0.01,itr=100):   
    cost_hist = np.zeros(itr)
    theta_hist = np.zeros((itr,2))
    for it in range(itr):
        pred = np.dot(X,theta)
        theta = theta - (1/m)*alpha*((X.T.dot(pred-Y)))
        theta_hist[it,:] = theta.T
        cost_hist[it] = cal_cost(m,theta,X,Y)
    return theta,cost_hist,theta_hist

alpha = 0.01
itr = 100
X = np.random.randn(300,1)
Y = np.random.randn(300,1)
theta = np.random.randn(2,1)
m = len(X)
vector_x = np.c_[np.ones([m,1]),X]
theta,cost_hist,theta_hist=grad_desc(m,theta,vector_x,Y,alpha,itr)
print('Theta0:  {:0.3f},\nTheta1   {:0.3f}'.format(theta[0][0],theta[1][0]))
print('MSE:  {:0.3f}'.format(cost_hist[-1]))
print('theta_hist')
plt.plot(theta_hist)
plt.ylabel('some numbers')
plt.show()
plt.plot(cost_hist)
plt.ylabel('some numbers')
plt.show()
print(theta_hist.shape)
print(Y.shape)
```

```python
#Implement Gradient Descent from scratch for Multivariate Linear Regression
import numpy as np
import matplotlib.pyplot as plt

def cal_cost(m,theta,X,Y):
    pred = np.dot(X,theta)
    cost = (1/(2*m))*np.sum(np.square(Y-pred))
    return cost


def grad_desc(m,theta,X,Y,alpha,itr):   
    cost_hist = np.zeros(itr)
    theta_hist = np.zeros((itr,4))
    for it in range(itr):
        pred = np.dot(X,theta)
        theta = theta - (1/m)*alpha*((X.T.dot(pred-Y)))
        theta_hist[it,:] = theta.T
        cost_hist[it] = cal_cost(m,theta,X,Y)
    return theta,cost_hist,theta_hist

alpha = 0.01
itr = 1000
X1 = np.random.randn(300,1)
X2 = np.random.randn(300,1)
X3 = np.random.randn(300,1)
Y = np.random.randn(300,1)
theta = np.random.randn(4,1)
m = len(X)
vector_x = np.c_[np.ones([m,1]),X1,X2,X3]

theta,cost_hist,theta_hist=grad_desc(m,theta,vector_x,Y,alpha,itr)
print('Theta0:  {:0.3f},\nTheta1   {:0.3f},\nTheta2   {:0.3f},\nTheta3   {:0.3f}'.format(theta[0][0],theta[1][0],theta[2][0],theta[3][0]))
print('MSE:  {:0.3f}'.format(cost_hist[-1]))
print('theta_hist')
plt.plot(theta_hist)
plt.ylabel('Theta')
plt.xlabel('Iteration')
plt.show()
plt.plot(cost_hist)
plt.ylabel('Cost Function')
plt.xlabel('Iteration')
plt.show()
print(theta_hist.shape)
print(Y.shape)
```
